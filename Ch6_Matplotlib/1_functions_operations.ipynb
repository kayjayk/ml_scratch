## Basic function & Operations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

#### jupyter 1 line에 figure가 있고 그 위에 plot을 쌓을 수 있다

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
plt.plot(range(100), range(100))
```

#### figure 반환도 가능

```python
fig = plt.figure() #figure 반환
fig.set_size_inches(10, 5) # 크기지정
ax_1 = fig.add_subplot(2,4,1) # row2개, column4개의 전체 plot인데 지금 그 중 첫번째 subplot
ax_2 = fig.add_subplot(2,2,4) # 4번째 subplot
ax_3 = fig.add_subplot(2,4,4)
ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show()
```

## Set color

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value+100 for value in X]

plt.plot(X_1, Y_1, color="#f43993")
plt.plot(X_2, Y_2, color="c")
plt.show()
```

## Set linestyle

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="c", ls='dotted')
```

## Set title

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="c", ls='dotted')
plt.title('$y = \\frac{ax + b}{test}$')
plt.show()
```

## Set legend

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line 1')
plt.plot(X_2, Y_2, c="c", ls='dotted', label='line 1')
plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.text(70, 70, "Line_abcd") # 앞의 숫자는 좌표
plt.annotate(
    'Line_xxx', xy=(50, 150), xytext=(20, 175), # 앞의 xy는 가리키는 곳, xytext는 화살표 시작지점
    arrowprops=dict(facecolor='r', shrink=0.125)) # shrink 화살표 수축정도.
plt.title('$y = ax+b$')
plt.show()
```

## Set grid & xylim

```python
X_1 = range(100000)
Y_1 = [value for value in X_1]
X_2 = range(100000)
Y_2 = [value+20000 for value in X_2]

plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line 1')
plt.plot(X_2, Y_2, c="c", ls='dotted', label='line 1')
plt.grid(True, lw=0.3, ls="--", c="gray")
plt.xlim(-1000, 200000)
plt.ylim(5000, 80000)
```

## Save plot

```python
X_1 = range(100000)
Y_1 = [value for value in X_1]
X_2 = range(100000)
Y_2 = [value+20000 for value in X_2]

plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line 1')
plt.plot(X_2, Y_2, c="c", ls='dotted', label='line 1')
plt.grid(True, lw=0.3, ls="--", c="gray")
plt.xlim(-1000, 200000)
plt.ylim(5000, 80000)
plt.savefig("test.png", c="a")
```
