## Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_1 = np.random.rand(512, 2) # 2차원 data가 512개
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:,0], # 전체 data중 0번째 (x축)
           data_1[:,1], # 전체 data중 1번째 (y축)
           c="b", marker="x") # 색상과 point 모양
plt.scatter(data_2[:,0], data_1[:,1], c="r", marker="^")

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5, marker='o') # alpha는 투명도
plt.show
```

## Bar chart

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)

plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
print(X)

data = np.array([[5., 5., 5., 5.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

for i in range(3):
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), 
            color = color_list[i], label=data_label[i])
plt.legend()
plt.show()
```

```python
A = [5., 30., 45., 22.]
B = [5, 25, 50, 20]

X = range(4)

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'r', bottom = 60)
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```
